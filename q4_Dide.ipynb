```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#importing pandas, seaborn and matplotlibs
```

```python
data = pd.read_csv('COVID1.csv')
#reading the file in csv format
```

```python
grouped_data = data.groupby(["Sector", "Company support level (1-10)"])["Productivity change (%)"].mean().reset_index()
#data is grouped to calculate the mean of the productivity change
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=grouped_data,
    x="Sector",
    y="Productivity change (%)",
    hue="Company support level (1-10)",
    palette="ch:s=.25,rot=-.25"
)
plt.title("Average Productivity Change by Sector and Company Support Level", fontsize=16, weight='bold')
plt.xlabel("Sector")
plt.ylabel("Average Productivity Change (%)")
plt.xticks(rotation=30, ha="right")
plt.legend(title="Company Support Level (1-10)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
#creating the bar plot
```
