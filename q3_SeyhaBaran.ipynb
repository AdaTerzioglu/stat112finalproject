```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("COVID1.csv")

sector_summary = df.groupby('Sector').agg({
   'Health impact (1-10)': 'mean',
}).reset_index()
sector_summary = sector_summary.sort_values('Health impact (1-10)', ascending=False).reset_index(drop=True)
plt.figure(figsize=(12, 6))
plt.stem(sector_summary.index,
          sector_summary['Health impact (1-10)'])

plt.xticks(sector_summary.index, sector_summary['Sector'])
plt.ylabel('Health impact')
plt.title('Health impacts By Sectors')

plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=0, linestyle='-')
plt.show()
```

```python
sector_summary1 = df.groupby('Sector').agg({
   'Stress level (1-10)': 'mean'
}).reset_index() 
sector_summary1 = sector_summary1.sort_values('Stress level (1-10)', ascending=True).reset_index(drop=True)
plt.figure(figsize=(12, 6))
plt.stem(sector_summary1.index,
          sector_summary1['Stress level (1-10)'])

plt.xticks(sector_summary1.index, sector_summary1['Sector'])
plt.ylabel('Stress level (1-10)')
plt.title('Stress level (1-10)  By Sectors')




plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=0, linestyle='-')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(sector_summary1.index,
          sector_summary1['Stress level (1-10)'], 'orange', label="Veri Seti 1")
plt.xticks(sector_summary1.index, sector_summary1['Sector'])
ax1.set_xlabel('Sectors')
ax1.set_ylabel('Stress level', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')
plt.xticks(fontsize=8)
ax2 = ax1.twinx()

ax2.plot(sector_summary.index,
          sector_summary['Health impact (1-10)'], 'b-', label="Veri Seti 2")
ax2.set_ylabel('Health impact', color='b')
ax2.tick_params(axis='y', labelcolor='b')
```
