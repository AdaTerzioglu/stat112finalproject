```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('COVIDfinall.csv')
```

```python
sector_order = ['finance','retail','technology','manufacturing', 'healthcare','education','other']
df['Sector'] = pd.Categorical(df['Sector'], categories=sector_order, ordered=True)


age_group_order = ['18-24', '25-34', '35-44', '45-54', '55+']
df['Age group'] = pd.Categorical(df['Age group'], categories=age_group_order, ordered=True)


heatmap_data = df.groupby(['Age group', 'Sector'], observed=False)['Stress level (1-10)'].mean().unstack()

plt.figure(figsize=(12, 8))  
sns.heatmap(heatmap_data, annot=False, cmap='Blues', fmt='.1f', annot_kws={"size": 12}, linewidths=0.5, cbar_kws={'label': 'Average Stress Level'})

plt.title('Average Stress Levels for Age Groups in Sectors', fontsize=18, weight='bold', color='darkblue')
plt.ylabel('Age Group', fontsize=14)
plt.xlabel('Sector', fontsize=14)

plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12, rotation=0)

plt.tight_layout()
plt.show()
```
