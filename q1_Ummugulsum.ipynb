```python
import pandas as pd   #import python libraries
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
 covid = "COVID1.csv"
```

```python
data = pd.read_csv(covid)
```

```python
data_long = pd.melt(data,
                    id_vars=['Sector'],
                    value_vars=['Health impact (1-10)', 'Job satisfaction (-5,5)'],
                    var_name='Metric',
                    value_name='Value')
plt.figure(figsize=(12, 6))
sns.violinplot(data=data_long,
               x='Sector',
               y='Value',
               hue='Metric',
               split=True,
               palette='muted')

plt.title("Health Impact and Job Satisfaction Distribution by Sector", fontsize=16)
plt.xlabel("Sector", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.tight_layout()
plt.show()
```
