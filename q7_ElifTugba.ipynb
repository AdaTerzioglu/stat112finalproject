```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reading Data
datam = pd.read_csv('kontrol.csv', encoding='ISO-8859-1', delimiter=',')

#setting style and context
sns.set_style("whitegrid") #adding a background to improve readability
sns.set_context("notebook") #adjusting the scale of the plot elements

# choosing and setting colors to use that I personally want
custom_palette =sns.color_palette(["#1E90FF", "#4682B4", "#87CEEB"])  # Various shades of blue
sns.set_palette(custom_palette)

# Create a Box Plot and adjust the size
plt.figure(figsize=(9.5,6)) #adjusting the plots size
graph = sns.boxplot(x="Sector", y="Income change (%)", hue="Gender", data=datam) #creating box plot 

# Customize the plot adding names of x, y axis and setting legend
plt.title("Change in Income (%) by Sector and Gender") # writing a title name
plt.xlabel("Sector") # giving x axis a name
plt.ylabel("Income change (%)")# giving y axis a name
plt.legend(title="Gender") #assigning legend
graph.legend (bbox_to_anchor = (1, 1)) # setting the position of the legend
# saving the plot as an image
plt.savefig("mygraph.png", dpi = 600) 
# Showing the plot
plt.show()
```
