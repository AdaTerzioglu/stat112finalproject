```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Read the COVID1 file
df = pd.read_csv("COVID1.csv", delimiter=',')

#Choose necessary columns and rename
plot_data = df[['Sector', 'Age group', 'Productivity change (%)']].copy()
plot_data.rename(columns={'Productivity change (%)': 'Productivity change'}, inplace=True)

#Set the size
plt.figure(figsize=(12, 8))

#Sort age groups
age_group_order = ["18-24", "25-34", "35-44", "45-54", "55+"]
plot_data['Age group'] = pd.Categorical(plot_data['Age group'], categories=age_group_order, ordered=True)

#Create 7 color palette of blue
blue_palette = sns.light_palette("blue", reverse=True, n_colors=7)

#Create the boxplot
sns.boxplot(
    data=plot_data,
    x="Age group",
    y="Productivity change",
    hue="Sector",
    palette=blue_palette,
    showfliers=True,
    dodge=True
)

#Legend setting
plt.legend(title="Sector", loc="upper left", bbox_to_anchor=(1, 1))

#Tite and tags
plt.title("Productivity Change Rates of Age Groups by Sector", fontsize=14)
plt.xlabel("Age group", fontsize=12)
plt.ylabel("Productivity change (%)", fontsize=12)

#Show the graph
plt.tight_layout()
plt.show()
```
