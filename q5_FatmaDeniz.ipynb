```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('COVID1.csv')  #reading the file


#creating the violin plot
sns.violinplot(
    data=data,
    x='Remote work',
    y='Productivity change (%)',
    order=['always', 'often', 'occasionally', 'rarely', 'never'],  #ordering according to the frequency
    palette='Blues'
)
plt.title('Productivity Change Distribution by Remote Work Frequency', fontsize=14)
plt.xlabel('Remote Work Frequency', fontsize=12)
plt.ylabel('Productivity Change (%)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

```
